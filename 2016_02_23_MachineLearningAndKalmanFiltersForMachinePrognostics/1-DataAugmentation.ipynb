```python
index_columns_names =  ["UnitNumber","Cycle"]
operational_settings_columns_names = ["OpSet"+str(i) for i in range(1,4)]
sensor_measure_columns_names =["SensorMeasure"+str(i) for i in range(1,22)]
input_file_column_names = index_columns_names + operational_settings_columns_names + sensor_measure_columns_names

train = pd.read_csv("train_FD004.txt", delim_whitespace=True, header=None)
train.columns = input_file_column_names
```

```python
number_original_units = train.UnitNumber.max()
train2 = train.copy()
train3 = train.copy()
train2.Cycle = train.Cycle + 1
train3.Cycle = train.Cycle - 1
train2.UnitNumber = train.UnitNumber + number_original_units
train3.UnitNumber = train.UnitNumber + 2 * number_original_units
train3 = train3[train3.Cycle > 0]
```

```python
train_augmented = train.append(train2).append(train3)
```

```python
train_augmented.to_csv("train_augmented_FD004.txt", sep=' ', header=False, index=False)
```
